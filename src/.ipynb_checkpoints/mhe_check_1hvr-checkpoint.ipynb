```julia
using DataFrames
using Gadfly
using Compose
set_default_plot_size(30cm, 12cm)
```

# 1hvr_apo

## minimization

```julia
dir_hvr_apo_min="/home/german/labo/16/exs/run/1hvr_apo/mhe/m"
hvr_apo_min_Etot = readdlm("$dir_hvr_apo_min/outputs/Etot_min");
hvr_apo_mmin_Etot = readdlm("$dir_hvr_apo_min/outputs/Etot_mmin");
df_hvr_apo_MIN_Etot = DataFrame(E=[hvr_apo_min_Etot[:, 2]; hvr_apo_mmin_Etot[:, 2]],
NSTEP=[hvr_apo_min_Etot[:, 1]; hvr_apo_mmin_Etot[:, 1]]);
```

```julia
plot(df_hvr_apo_MIN_Etot, Geom.line, x="NSTEP", y="E",
Coord.Cartesian(xmin=0, xmax=100),
Guide.title("Solvent minimization + solute minimization"),
Theme(minor_label_font_size = 10pt, major_label_font_size = 15pt, key_title_font_size = 30pt))
```

## heat

```julia
# Energies
dir_hvr_apo_hea="/home/german/labo/16/exs/run/1hvr_apo/mhe/h"
hvr_apo_hea_Ek = readdlm("$dir_hvr_apo_hea/outputs/ektot.dat");
hvr_apo_hea_Ep = readdlm("$dir_hvr_apo_hea/outputs/eptot.dat");
hvr_apo_hea_Etot = readdlm("$dir_hvr_apo_hea/outputs/etot.dat");
df_hvr_apo_HEA_E = DataFrame(E=[hvr_apo_hea_Ek[:, 2]; hvr_apo_hea_Ep[:, 2]; hvr_apo_hea_Etot[:, 2]],
NSTEP=[hvr_apo_hea_Ek[:, 1]; hvr_apo_hea_Ep[:, 1]; hvr_apo_hea_Etot[:, 1]];
Type=repeat(["Ek"; "Ep"; "Etot"], inner=[length(hvr_apo_hea_Ek[:, 1])]));

# Temperature
hvr_apo_hea_T = readdlm("$dir_hvr_apo_hea/outputs/temp.dat");
df_hvr_apo_HEA_T = DataFrame(Temperature=hvr_apo_hea_T[:, 2],
NSTEP=hvr_apo_hea_T[:, 1]);
```

```julia
plot(df_hvr_apo_HEA_E, Geom.line, x="NSTEP", y="E", color="Type",
Guide.title("Heating"),
Theme(minor_label_font_size = 10pt, major_label_font_size = 15pt,
key_title_font_size=20pt, key_label_font_size=20pt))
```

```julia
plot(df_hvr_apo_HEA_T, Geom.line, x="NSTEP", y="Temperature",
Guide.title("Heating"),
Theme(minor_label_font_size = 10pt, major_label_font_size = 15pt,
key_title_font_size=20pt, key_label_font_size=20pt))
```

## equilibration

```julia
# Energies
dir_hvr_apo_equ="/home/german/labo/16/exs/run/1hvr_apo/mhe/e"
hvr_apo_equ_Ek = readdlm("$dir_hvr_apo_equ/outputs/ektot.dat");
hvr_apo_equ_Ep = readdlm("$dir_hvr_apo_equ/outputs/eptot.dat");
hvr_apo_equ_Etot = readdlm("$dir_hvr_apo_equ/outputs/etot.dat");
df_hvr_apo_EQU_E = DataFrame(E=[hvr_apo_equ_Ek[:, 2]; hvr_apo_equ_Ep[:, 2]; hvr_apo_equ_Etot[:, 2]],
NSTEP=[hvr_apo_equ_Ek[:, 1]; hvr_apo_equ_Ep[:, 1]; hvr_apo_equ_Etot[:, 1]];
Type=repeat(["Ek"; "Ep"; "Etot"], inner=[length(hvr_apo_equ_Ek[:, 1])]));

# Temperature
hvr_apo_equ_T = readdlm("$dir_hvr_apo_equ/outputs/temp.dat");
df_hvr_apo_EQU_T = DataFrame(Temperature=hvr_apo_equ_T[:, 2],
NSTEP=hvr_apo_equ_T[:, 1]);

# RMSD ntwx=500
hvr_apo_equ_rmsd = readdlm("$dir_hvr_apo_equ/rmsd_equ_1hvr_apo");
df_hvr_apo_EQU_RMSD = DataFrame(RMSD=hvr_apo_equ_rmsd[:, 2], Frame=hvr_apo_equ_rmsd[:, 1]);
```

```julia
plot(df_hvr_apo_EQU_E, Geom.line, x="NSTEP", y="E", color="Type",
Guide.title("Equilibration"),
Theme(minor_label_font_size = 10pt, major_label_font_size = 15pt,
key_title_font_size=20pt, key_label_font_size=20pt))
```

```julia
plot(df_hvr_apo_EQU_T, Geom.line, x="NSTEP", y="Temperature",
Guide.title("Equilibration"),
Theme(minor_label_font_size = 10pt, major_label_font_size = 15pt,
key_title_font_size=20pt, key_label_font_size=20pt))
```

```julia
plot(df_hvr_apo_EQU_RMSD, Geom.line, x="Frame", y="RMSD",
Guide.title("Equilibration"),
Guide.annotation(compose(compose(context(), text(2500, 1.3, "frame = 1ps")), fontsize(15pt), fill("black"))),
Theme(minor_label_font_size = 10pt, major_label_font_size = 15pt,
key_title_font_size=20pt, key_label_font_size=20pt))
```
